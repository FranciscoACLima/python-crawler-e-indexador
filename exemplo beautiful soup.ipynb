```python
# instalação das bibliotecas
# pip install -r requeriments.txt
from bs4 import BeautifulSoup
import urllib3
```

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
http = urllib3.PoolManager()
pagina = http.request('GET', 'http://www.iaexpert.com.br')
# pagina = http.request('GET', 'https://ovoodaserpenteemplumada.com')
# pagina = http.request('GET', 'https://pt.wikipedia.org/wiki/Linguagem_de_programa%C3%A7%C3%A3o')

pagina.status
```

```python
sopa = BeautifulSoup(pagina.data, "lxml")
sopa
```

```python
sopa.title
```

```python
sopa.title.string
```

```python
links = sopa.find_all('a')
len(links)
```

```python
for link in links:
    print(link.get('href'))
    print(link.contents)
    print('--------------------------------')
```
